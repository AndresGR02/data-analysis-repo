## Estimación de los niveles de obesidad basado en hábitos de alimentación y condición física

**Columnas:**

*Gender:*
Genéro

*Age:*
Edad

*Height:*
Altura

*Weight:*
Peso

*family_history_with_overweight:*
Tienes algún familiar que sufrió o sufre sobrepeso?

*FAVC:*
Sueles comer alimentos con altas calorías?

*FCVC:*
Sueles comer vegetales en tus comidas?

*NCP:*
Cuántas comidas principales tienes al día?

*CAEC:*
Comes alguna comida entre las comidas principales del día?

*SMOKE:*
Fumas?


*CH2O:*
Cuanta agua bebes diariamente?

*SCC:*
Monitoreas la cantidad de calorías que consumes diariamente?

*FAF:*
Qué tan seguido haces actividades físicas?

*TUE:*
Qué tanto tiempo usas dispositivos tecnológicos?

*CALC:*
Qué tan seguido bebes alcohol?

*MTRANS:*
Qué medio de transporte utilizas generalmente?

*NObeyesdad:*
Nivel de obesidad


**Link del CSV:**
https://archive.ics.uci.edu/dataset/544/estimation+of+obesity+levels+based+on+eating+habits+and+physical+condition

```python
import pandas as pd
from IPython.display import HTML

# Load data
datos = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv', delimiter=',', decimal='.', header=0)

# Convert DataFrame to HTML
result = datos.to_html(classes='table table-striped')

# Create HTML object
table = HTML(result)

display(table)
#datos.sample(5)

```

```python
print("El maximo de edad es: ")
print(datos['Age'].max(),'\n')
print("El minimo de edad es: ")
print(datos['Age'].min(),'\n')
print("La suma de las edades es: ")
print(datos['Age'].sum(),'\n')
print("Los datos ordenados descendentemente: ")
print(datos.sort_values(by='Age', ascending=False),'\n')
print("El resultado de filtrar la columna ¿Cuanta agua bebes diariamente? por el valor 2.0: ")
print(datos.query('CH2O==2.0'),'\n')
```
